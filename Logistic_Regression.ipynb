### Missing value treatment

```python
df['age'].fillna(df['age'].mean(), inplace=True)
df['embark_town'].fillna('Southampton',inplace=True)
df.isnull().mean() * 100
```

```python
df.head()
```

```python
df.drop('deck', axis=1, inplace=True)
df.drop('who', axis=1, inplace=True)
df.drop('pclass', axis=1, inplace=True)
df.drop('alive', axis=1, inplace=True)
df.drop('embarked', axis=1, inplace=True)
df.head()
```

```python
df.isnull().mean() * 100
```

```python
import pandas as pd
df = pd.get_dummies(df, columns=['sex'], drop_first=True, prefix='sex')
df = pd.get_dummies(df, columns=['class'], drop_first=True, prefix='class')
df = pd.get_dummies(df, columns=['adult_male'], drop_first=True, prefix='adult_male')
df = pd.get_dummies(df, columns=['embark_town'], drop_first=True, prefix='embark_town')
df = pd.get_dummies(df, columns=['alone'], drop_first=True, prefix='alone')
df.describe()
```

```python
df.columns
```

```python
target_column = ['survived']
predictors = list(set(list(df.columns))-set(target_column))
df[predictors] = df[predictors]/df[predictors].max()
df.describe()
```

### Simple Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Define the independent variable (X) and dependent variable (y)
X = df[['age']]
y = df['survived']

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

# Create a logistic regression model
model = LogisticRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

# Predict the target variable on train and test sets
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Compute the confusion matrix for train and test sets
confusion_train = confusion_matrix(y_train, y_train_pred)
confusion_test = confusion_matrix(y_test, y_test_pred)

# Compute accuracy, sensitivity, specificity, and F1 score for train and test sets
accuracy_train = accuracy_score(y_train, y_train_pred)
accuracy_test = accuracy_score(y_test, y_test_pred)

sensitivity_train = recall_score(y_train, y_train_pred)
sensitivity_test = recall_score(y_test, y_test_pred)

specificity_train = confusion_train[0, 0] / (confusion_train[0, 0] + confusion_train[0, 1])
specificity_test = confusion_test[0, 0] / (confusion_test[0, 0] + confusion_test[0, 1])

f1_train = f1_score(y_train, y_train_pred)
f1_test = f1_score(y_test, y_test_pred)

# Print the confusion matrix and evaluation metrics for train and test sets
print("Confusion Matrix - Train Set:")
print(confusion_train)
print("\nConfusion Matrix - Test Set:")
print(confusion_test)

print("\nAccuracy - Train Set:", accuracy_train)
print("Accuracy - Test Set:", accuracy_test)

print("\nSensitivity - Train Set:", sensitivity_train)
print("Sensitivity - Test Set:", sensitivity_test)

print("\nSpecificity - Train Set:", specificity_train)
print("Specificity - Test Set:", specificity_test)

print("\nF1 Score - Train Set:", f1_train)
print("F1 Score - Test Set:", f1_test)
```
